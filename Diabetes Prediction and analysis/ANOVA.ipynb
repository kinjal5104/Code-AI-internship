```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import os
```

```python
url = "https://raw.githubusercontent.com/kuchbhi-kunal/nidan/main/diabetes_binary_5050split_health_indicators_BRFSS2015.csv"
df = pd.read_csv(url)
df.head()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f_oneway

# Assume df is your DataFrame containing the data

# Fit ANOVA model
model = ols('HighBP ~ C(Diabetes_binary)', data=df).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
print(aov_table)

# Interpret the result
alpha = 0.05
if aov_table['PR(>F)']['C(Diabetes_binary)'] < alpha:
    print('Reject null hypothesis. There are significant differences in high blood pressure with and without diabetes.')
else:
    print('Fail to reject null hypothesis. There are no significant differences in high blood pressure with and without diabetes.')

# Separate data based on Diabetes_binary
sqft_living_no_waterfront = df[df['Diabetes_binary'] == 0]['HighBP']
sqft_living_waterfront = df[df['Diabetes_binary'] == 1]['HighBP']

# Perform ANOVA
f_statistic, p_value = f_oneway(sqft_living_no_waterfront, sqft_living_waterfront)

print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpret the results
if p_value < alpha:
    print('Reject null hypothesis. There are significant differences in high blood pressure with and without diabetes.')
else:
    print('Fail to reject null hypothesis. There are no significant differences in high blood pressure with and without diabetes.')
```
